# Simple Regression 

## Solution 1 --> closed-form solution

```python
x = np.array([0,1,2,3,4])
y = np.array([1,3,7,13,21])
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
xy_mean = np.mean(x*y)
x_square_mean = np.mean(x**2)
```

```python
w1 = (x_mean*y_mean - xy_mean)/(x_mean**2 - x_square_mean)
w0 = (y_mean - w1*x_mean)
```

```python
print("w0 :", w0, "\n w1 :", w1)
```
